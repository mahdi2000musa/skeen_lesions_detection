```python
from glob import glob
from keras.preprocessing import image
import numpy as np
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from catboost import CatBoostClassifier
```

```python
import pandas as pd

# reading train file

df = pd.read_csv('/content/drive/MyDrive/ISIC-2017_Training_Part3_GroundTruth (2).csv')
train_data_name = df.iloc[:,0].values
melanomas = df.iloc[:,1].values
seb = df.iloc[:,-1].values
```

```python
#reading test file
df = pd.read_csv('/content/drive/MyDrive/ISIC-2017_Test_v2_Part3_GroundTruth (1).csv')
test_data_name = df.iloc[:,0].values
test_melanomas = df.iloc[:,1].values
test_seb = df.iloc[:,-1].values
```

```python
# getting photo path and divide in 3 class

test_path = {

    'seborrheic_keratosis' : [],
    'benign_nevi' : []
}

for x in range(600):
  path = '/content/drive/MyDrive/ISIC-2017_Test_v2_Data/' + test_data_name[x] + '.jpg'

  if test_melanomas[x] == 0 and test_seb[x] :
    test_path['seborrheic_keratosis'].append(path)
  elif test_melanomas[x] == 0 and test_seb[x] == 0:
    test_path['benign_nevi'].append(path)

len(test_path['seborrheic_keratosis'])
```

```python
train_path = {

    'seborrheic_keratosis' : [],
    'benign_nevi' : []
}

for x in range(2000):
  path = '/content/drive/MyDrive/ISIC-2017_Training_Data/' + train_data_name[x] + '.jpg'
  if melanomas[x] == 0 and seb[x] :
    train_path['seborrheic_keratosis'].append(path)
  elif melanomas[x] == 0 and seb[x] == 0:
    train_path['benign_nevi'].append(path)

len(train_path['seborrheic_keratosis'])
```

```python
image_classes = {

    'seborrheic_keratosis': 0,
    'benign_nevi': 1
}
```

```python
# resize the test image and add to list
import cv2

x_te = []
y_te = []
counter = 0
for lable in test_path:
  for image_path in test_path[lable]:
    print(counter, '/' , all)
    counter += 1
    img = image.load_img(image_path, target_size=(224,224))
    x_te.append(img)
    y_te.append(image_classes[lable])
```

```python
# resize and add to list train photos

x_tr = []
y_tr = []
counter = 0
for lable in train_path:
  for image_path in train_path[lable]:
    print(counter, '/' , all)
    counter += 1
    img = image.load_img(image_path, target_size=(224,224))
    x_tr.append(img)
    y_tr.append(image_classes[lable])
```

```python
# convert train list to np array

x_trains= []
y_train = np.array(y_tr)

for xitem  in x_tr :
  x_trains.append(np.array(xitem)/255.0)

x_train = np.array(x_trains)


```

```python
# convert test list to np array

x_tests= []
y_test = np.array(y_te)

for xitem  in x_te :
  x_tests.append(np.array(xitem)/255.0)

x_test = np.array(x_tests)
```

```python
#add pretrained model

from tensorflow.keras.applications.densenet import  DenseNet201
from keras.models import Model
import keras

pre_trained1 = DenseNet201(include_top = False, input_shape = (224,224,3), pooling='avg', weights = 'imagenet')
```

```python
#add pretrained model

from tensorflow.keras.applications.vgg16 import  VGG16
from keras.models import Model
import keras

pre_trained2 = VGG16(include_top = False, input_shape = (224,224,3), pooling='avg', weights = 'imagenet')
```

```python

x_train = np.append(x_train1, x_train2, axis=1)
x_test = np.append(x_test1,x_test2, axis=1)

print(x_train.shape, x_test.shape)
```

```python
accs = 0


classifier = CatBoostClassifier(learning_rate = 0.2, iterations= 125,random_state = 0, depth = 7,silent = True)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
accs= accuracy_score(y_test, y_pred)
print(accs)
```
